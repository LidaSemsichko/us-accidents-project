```python
%sh
# перевіримо, що все ок
git status

# додаємо файли
git add .

# створюємо коміт
git commit -m "Initial commit from Databricks"

# називаємо гілку main (може сказати, що вона вже існує — це не страшно)
git branch -M main

# додаємо remote (якщо скаже 'remote origin already exists', тоді використай дві останні команди нижче)
git remote add origin https://github.com/LidaSemsichko/us-accidents-project.git

# якщо раптом було 'remote origin already exists', тоді зроби замість add:
# git remote set-url origin https://github.com/LidaSemsichko/us-accidents-project.git

# пушимо
git push -u origin main
```
